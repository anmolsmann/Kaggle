```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
train =pd.read_csv("trainc.csv")
test = pd.read_csv("testc.csv")
real = pd.read_csv("cirrhosis.csv")
```

```python
test_id = test['id']
```

```python
train.info()
```

```python
test.info()
```

```python
real.info()
```

```python
real.dropna(inplace=True)
```

```python
train = pd.concat([train,real]).reset_index(drop=True)
```

```python
train.drop(columns=['id','ID'],inplace=True)
```

```python
test.drop(columns='id',inplace=True)
```

```python
train.head()
```

```python
test.head()
```

```python
#Encoding the Status column
target_mapping = { 'D':0,'C':1,'CL':2}
train['Status'] = train['Status'].map(target_mapping)
```

```python
#Encoding the values
categorical_cols = ['Drug','Sex','Ascites','Hepatomegaly','Spiders','Edema','Stage']
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
for i in categorical_cols:
    train[i] = encoder.fit_transform(train[i])
    test[i] = encoder.transform(test[i])
```

```python
train.head()
```

```python
test.head()
```

```python
 y = train['Status']
```

```python
train.drop(columns='Status',inplace=True)
```

```python
#Scaling the data using Standard Scaler
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X = scaler.fit_transform(train)
X = pd.DataFrame(X)
X.columns = train.columns

test_scaled = scaler.transform(test)
test_scaled = pd.DataFrame(test_scaled)
test_scaled.columns = test.columns
```

```python
X.head()
```

```python
test_scaled.head()
```

```python
#For the model i will be using XGBclassifier with parameters tuned from optuna
from xgboost import XGBClassifier
xg_params = {'n_estimators': 528,
 'learning_rate': 0.06542605284995028,
 'max_depth': 5,
 'subsample': 0.7109369692740082,
 'colsample_bytree': 0.13027270667695215,
 'min_child_weight': 1}

model = XGBClassifier(**xg_params)
```

```python
model.fit(X,y)
```

```python
preds = model.predict_proba(test_scaled)
```

```python
assert preds.shape == (test_scaled.shape[0],3)
submission_labels = ["Status_D", "Status_C", "Status_CL"]
sub =  pd.DataFrame({"id": test_id, **dict(zip(submission_labels, preds.T))})
```
